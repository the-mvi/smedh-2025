## Основные теоретические положения

Если случайны величины независимы, то $f(x,y)=f_1(x)\cdot f_2(y)$ иначе между ними может быть:

1) функциональная зависимость:
$y = g(x)$

2) статистическая зависимость:
$\phi (x|y) = f(x,y)/f_2(y)$

$\phi (y|x) = f(x,y)/f_1(x)$

Корелляционная зависимость - статистичеакая зависимость.

**Коэффициент корелляции**
$$r_{xy} = \frac{\mu_{xy}}{\sigma_x \sigma_y},$$
для него справедливо: $|r_{xy}|\le 1$'

Справедливы утверждения:

X и Y кореллированы -> СВ X и Y зависимы

X и Y независимы -> СВ X и Y некореллированы

### Пункт 4.
Исходя из результатов корреляционной таблицы вычислить значение выборочного коэффициента корреляции двумя способами: с помощью стандартной формулы и с помощью условных вариант. Убедиться, что результаты совпадают. Сделать выводы.

$\bar r_{xy} = \frac{\sum_{i=1}^{K_y} \sum_{j=1}^{K_x} n_{ij}y_ix_j - N\bar x_в \bar y_в}{NS_xS_y}$

```python
intervals1 = bins1.cat.categories
intervals2 = bins2.cat.categories

n_x = np.array(contingency_table_use.sum(axis=0))
n_y = np.array(contingency_table_use.sum(axis=1))

x = np.array([(interval.left + interval.right) / 2 for interval in intervals1])
y = np.array([(interval.left + interval.right) / 2 for interval in intervals2])

contingency_table = np.array(contingency_table_use)
N = np.sum(contingency_table)

x_mean = np.sum(n_x*x)/N
D_x = 1/N*(np.sum(n_x*(x - x_mean)**2))
S_x = np.sqrt((N-1)/N*D_x)

y_mean = np.sum(n_y*y)/N
D_y = 1/N*(np.sum(n_y*(y - y_mean)**2))
S_y = np.sqrt((N-1)/N*D_y)

sum = 0
for i in range(len(contingency_table[0])):
  for j in range(len(contingency_table[1])):
    sum += contingency_table[i,j]*x[i]*y[j]

cov = (sum - N*x_mean*y_mean)/N

r_xy = cov/(S_x*S_y)
print(r_xy)
```

```python
C_x = x[np.argmax(n_x)]
C_y = y[np.argmax(n_y)]

h_x = (df["Weight"].max() - df["Weight"].min())/k
h_y = (df["NCP"].max() - df["NCP"].min())/k

u = (x-C_x)/h_x
v = (y-C_y)/h_y

u_mean = np.sum(u*n_x)/N
v_mean = np.sum(v*n_y)/N

u2_mean = np.sum(u**2*n_x**2)/N
v2_mean = np.sum(v**2*n_y**2)/N

sigma_u = np.sqrt(u2_mean - u_mean**2)
sigma_v = np.sqrt(v2_mean - v_mean**2)

sum = 0
for i in range(len(contingency_table[0])):
  for j in range(len(contingency_table[1])):
    sum += contingency_table[i,j]*u[i]*v[j]

cov = (sum - N*u_mean*v_mean)/N

r_xy = cov/(sigma_u*sigma_v)
print(r_xy)
```

Значение выборочного коэффициента корреляции вычислено двумя способами:
С помощью стандартной формулы и с помощью условных вариант. Оба метода дают одинаковый результат: $r_{xy}\approx 0.104$. Низкое значение коэффициента корреляции указывает на слабую линейную зависимость между признаками первой и второй выборки.

### Пункт 5.
Построить доверительный интервал для коэффициента корреляции при уровне значимости
γ
∈
{
0.95
,
0.99
}
, сделать выводы.

```python
import scipy.stats as stats

def confidence_interval_r(r_xy, N, gamma):
    z = 0.5 * np.log((1 + r_xy)/(1 - r_xy))
    alpha = 1 - gamma
    _lambda = stats.norm.ppf(1 - alpha / 2)

    left  = z - _lambda / np.sqrt(N - 3)
    right = z + _lambda / np.sqrt(N - 3)

    left  = (np.exp(2 * left) - 1) / (np.exp(2 * left) + 1)
    right = (np.exp(2 * right) - 1) / (np.exp(2 * right) + 1)

    return left, right

left_95, right_95 = confidence_interval_r(r_xy, N, 0.95)
left_99, right_99 = confidence_interval_r(r_xy, N, 0.99)

print(f"Доверитиельный интервал r: ({left_95:.4f}; {right_95:.4f})")
print(f"Доверитиельный интервал r: ({left_99:.4f}; {right_99:.4f})")
```

Интервалы содержат ноль, что подтверждает гипотезу о возможном отсутствии значимой корреляции между признаками.Широкий диапазон интервалов объясняется небольшим объемом выборки N=95. Доверительный интеравал при уровне значимости $\gamma = 0.99$ ожидаемо оказался больше.

### Пункт 6.
Осуществить проверку статистической гипотезы о равенстве коэффициента корреляции нулю при заданном уровне значимости
α
=
0.05
, сделать выводы.

```python
# проверка гипотезы
alpha = 0.05
t_sample = r_xy * np.sqrt(N-2) / np.sqrt(1 - r_xy**2)
t_crit = stats.t.ppf(1-alpha, N-2)

print(f't выборочное  = {t_sample:.4f}')
print(f't критическое = {t_crit:.4f}')
if t_sample <= t_crit:
    print('H0 не отвергается')
else:
    print('Отвергаем гипотезу H0')
```

Поскольку $t_{выб} <t_{крит}$ , нет оснований отвергать нулевую гипотезу.Таким образом, можно сделать вывод, что коэффициент корреляции статистически не отличается от нуля.
